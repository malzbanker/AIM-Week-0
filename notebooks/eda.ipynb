```python
sns.histplot(dfb['GHI'], kde=True, bins=20, color='blue')
plt.title("Distribution of GHI")
plt.xlabel("GHI")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot for Petal Width by Species
sns.boxplot(data=dfb, x='GHI', y='DNI', palette='pastel')
plt.title("GHI by DNI")
plt.xlabel("GHI")
plt.ylabel("DNI")
plt.show()
```
